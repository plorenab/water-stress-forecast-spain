```python
import requests
import zipfile
import io
import subprocess
import os

import pandas as pd
```

```python
# Definir urls y rutas
zip_url = "https://www.miteco.gob.es/content/dam/miteco/es/agua/temas/evaluacion-de-los-recursos-hidricos/boletin-hidrologico/Historico-de-embalses/BD-Embalses.zip"
zip_path = "../data/BD-Embalses.zip"
extract_dir = "../data/embalses_data"
mdb_path = "../data/embalses_data/BD-Embalses.mdb"
csv_path = "../files/embalses_historico.csv"
tabla_principal = "T_Datos Embalses 1988-2026" 
```

```python
# Descargar el ZIP
print("Descargando ZIP desde el ministerio...")
r = requests.get(zip_url)
with open(zip_path, "wb") as f:
    f.write(r.content)
print("ZIP descargado correctamente")
```

```python
# Extraer el .mdb
print("Extrayendo .mdb...")
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_dir)
print(f".mdb extraído en '{extract_dir}'")
```

```python
# Exportar la tabla a CSV
os.makedirs(os.path.dirname(csv_path), exist_ok=True)
print(f"Exportando tabla a CSV en {csv_path} ...")
with open(csv_path, "w") as f:
    subprocess.run(["mdb-export", mdb_path, tabla_principal], stdout=f, check=True)
print(f"Tabla exportada a '{csv_path}'")
```
